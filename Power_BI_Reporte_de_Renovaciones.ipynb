Fuente: BBDD empresa B2B (datos censurados)

El siguiente reporte fue realizado para consolidar toda la informacion relacionada con la renovacion de clientes.

1. Resumen RNV permite visualizar las carteras activas asignadas por cantidad y monto mostrando el avance de renovados vs meta y por renovar, ademas permite ver la composicion de estados de renovacion, detalle de asignacion por ejecutivos.
Los filtros por Region, Ejecutivos, tipo de cliente y tipo de cartera

image.png

El reporte de aignacion de cartera permite visualizar de forma puntual aspectos coo la asigancion diaria del mes filtrado, el desglose por ejecutivos y desglose por estados, permite tener el analisis de como fueron las renovaciones entre carteras asi sean pasadas y como va actualizandose las carteras a futuro.

image.png

Desglose de cartera por region permite visualizar el total por renovar por region considerando avances de la renovacion en forma de clientes en negociacion lo que permite realizar proyecciones mensuales de cierre tanto por cantidad de planes como monto de venta por plan., los filtros disponibles son ejecutivo y recurrencia del cliente.

image.png

El reporte de negociacion especial permite controlar el monto ofrecido como descuento para evitar el churn de algunos clientes estrategicos tanto por cantidad de clientes como cantidad en monto donde se puede evideciar los planes, las regiones, algunos kpis como descuento total promedio y ticket promedio. Mantener el control de los descuentos que se den es vital para mantener saludable el negocio.

image.png

Del mismo modo medir la cantidad de descuentos ofrecidos por ejecutivos prmite medir la capacidad de gestion de leads para establecer correcciones o ajustar estrategias en periodos puntuales.
